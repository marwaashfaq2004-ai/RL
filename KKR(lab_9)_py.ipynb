```python
def check_robot_navigation(room_relation):
    """
    Checks if a robot can navigate between rooms based on their RCC relation.
    'EC' = Externally Connected
    'DC' = Disconnected
    """
    if room_relation == "EC":
        return "Navigation Allowed: The rooms share a boundary/doorway."

    elif room_relation == "DC":
        return "Navigation Blocked: The rooms are completely disconnected."

    else:
        return "Unknown spatial relationship."


# --- PART 2: TEMPORAL REASONING (Interval Overlap Checker) ---

def check_schedule_overlap(event1_start, event1_end,
                           event2_start, event2_end):
    """
    Determines whether two scheduled events overlap.
    """

    if event1_end <= event2_start or event2_end <= event1_start:
        return "No Overlap: Events are scheduled safely."

    else:
        return "Conflict Detected: The event intervals overlap!"


# --- MAIN PROGRAM ---

if __name__ == "__main__":

    print("=== Spatial Reasoning Test ===")

    # Room A connected to Hallway H
    print("Room A to Hallway H:",
          check_robot_navigation("EC"))

    # Room C disconnected from Hallway H
    print("Room C to Hallway H:",
          check_robot_navigation("DC"))

    print("\n=== Temporal Reasoning Test ===")

    # Scenario 1
    print("Schedule 1:",
          check_schedule_overlap(14, 16, 16, 17))

    # Scenario 2
    print("Schedule 2:",
          check_schedule_overlap(17, 19, 18, 20))
```
