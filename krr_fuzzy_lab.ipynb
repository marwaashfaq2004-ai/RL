```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
# 1. Define the Variables (Inputs and Outputs)
temperature = ctrl.Antecedent(np.arange(0, 41, 1), 'temperature')  # 0 to 40°C
fan_speed = ctrl.Consequent(np.arange(0, 101, 1), 'fan_speed')     # 0% to 100%
```

```python
# 2. Define Membership Functions (Triangular)
temperature['cold'] = fuzz.trimf(temperature.universe, [0, 0, 20])
temperature['warm'] = fuzz.trimf(temperature.universe, [15, 25, 35])
temperature['hot'] = fuzz.trimf(temperature.universe, [30, 40, 40])
fan_speed['slow'] = fuzz.trimf(fan_speed.universe, [0, 0, 50])
fan_speed['fast'] = fuzz.trimf(fan_speed.universe, [40, 100, 100])
```

```python

# 3. Define the Knowledge Base (Fuzzy Rules)
rule1 = ctrl.Rule(temperature['cold'], fan_speed['slow'])
rule2 = ctrl.Rule(temperature['hot'], fan_speed['fast'])
rule3 = ctrl.Rule(temperature['warm'], fan_speed['fast'])
```

```python
# 4. Create and Simulate the Control System
fan_control = ctrl.ControlSystem([rule1, rule2, rule3])
fan_simulator = ctrl.ControlSystemSimulation(fan_control)

# Pass a crisp test value to the system
test_temp = 34
fan_simulator.input['temperature'] = test_temp

# Perform the fuzzy inference processing
fan_simulator.compute()

# Output the defuzzified crisp result
print(f"Simulated Room Temperature: {test_temp}°C")
print(f"Calculated Target Fan Speed: {fan_simulator.output['fan_speed']:.2f}%")
```

```python


import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# 1. Define Input and Output Variables
occupancy = ctrl.Antecedent(np.arange(0, 11, 1), 'occupancy')  # 0 to 10 people
cooling_power = ctrl.Consequent(np.arange(0, 101, 1), 'cooling_power')  # 0% to 100%

# 2. Define Membership Functions
occupancy['low'] = fuzz.trimf(occupancy.universe, [0, 0, 5])
occupancy['high'] = fuzz.trimf(occupancy.universe, [4, 10, 10])

cooling_power['low_power'] = fuzz.trimf(cooling_power.universe, [0, 0, 50])
cooling_power['high_power'] = fuzz.trimf(cooling_power.universe, [50, 100, 100])

# 3. Define Fuzzy Rules
rule1 = ctrl.Rule(occupancy['low'], cooling_power['low_power'])
rule2 = ctrl.Rule(occupancy['high'], cooling_power['high_power'])

# 4. Create Control System
ac_control = ctrl.ControlSystem([rule1, rule2])
ac_simulator = ctrl.ControlSystemSimulation(ac_control)

# 5. Test with 7 people
test_occupancy = 7
ac_simulator.input['occupancy'] = test_occupancy

# Perform fuzzy inference
ac_simulator.compute()

# Display result
print(f"Room Occupancy: {test_occupancy} people")
print(f"Cooling Power: {ac_simulator.output['cooling_power']:.2f}%")
```
