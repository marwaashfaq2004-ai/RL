```python
def infer_temperature_trend(heater_on: bool, window_open: bool) -> str:
    # If heater is OFF, temperature remains stable
    if not heater_on:
        return "stable"

    # Exception rule: Open window overrides default assumption
    if window_open:
        return "decreasing"

    # Default rule: Heater ON means temperature increases
    return "increasing"


# Test Case 1: Default reasoning applies
print(f"Heater ON, Window CLOSED -> Temperature is: "
      f"{infer_temperature_trend(heater_on=True, window_open=False)}")

# Test Case 2: Exception introduced
print(f"Heater ON, Window OPEN -> Temperature is: "
      f"{infer_temperature_trend(heater_on=True, window_open=True)}")
```
